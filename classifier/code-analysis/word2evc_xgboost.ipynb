```
import MeCab
import subprocess

cmd='echo `mecab-config --dicdir`"/mecab-ipadic-neologd"'
path = (subprocess.Popen(cmd, stdout=subprocess.PIPE,
                           shell=True).communicate()[0]).decode('utf-8')
m_t = MeCab.Tagger("-Owakati  -d {0}".format(path))
```

```
# -*- coding: utf-8 -*-
import os
import urllib.request
from collections import Counter

from gensim import corpora


def maybe_download(path):
    url = 'http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt'
    if os.path.exists(path):
        print('File already exists.')
    else:
        print('Downloading...')
        # Download the file from `url` and save it locally under `file_name`:
        urllib.request.urlretrieve(url, path)

def remove_stopwords(words, stopwords):
    words = [word for word in words if word not in stopwords]
    return words

def create_stopwords(file_path):
    stop_words = []
    for w in open(path, "r"):
        w = w.replace('\n','')
        if len(w) > 0:
          stop_words.append(w)
    return stop_words

path = "./data/stopwards.txt"
maybe_download(path)
stopwords = create_stopwords(path)
```

```
def tokenizer_mecab(text):
    text = m_t.parse(text)  # これでスペースで単語が区切られる
    ret = text.strip().split()  # スペース部分で区切ったリストに変換
    return ret

# 前処理として正規化をする関数を定義
import re

# cleaning & normarlize
def preprocessing_text(text):
    # 改行、半角スペース、全角スペースなどを削除
    text = re.sub('\r', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('　', '', text)
    text = re.sub(' ', '', text)
    text = re.sub('_', '', text)
    text = re.sub('＿', '', text)
    text = re.sub('@', '', text)
    text = re.sub('▁', '', text)

    # 数字文字の一律「0」化
    text = re.sub(r'[0-9 ０-９]', '0', text)  # 数字
    text = text.lower()

    return text

def tokenizer_with_preprocessing(text):
    text = preprocessing_text(text)  # 前処理の正規化
    words = tokenizer_mecab(text)  # Mecabの単語分割
    ret = remove_stopwords(words, stopwords) #stopwords
    return ret
```

```
import torchtext

# tsvやcsvデータを読み込んだときに、読み込んだ内容に対して行う処理を定義
# 文章とラベルの両方に用意します

max_length = 25
TEXT = torchtext.data.Field(sequential=True, tokenize=tokenizer_with_preprocessing,
                            use_vocab=True, lower=True, include_lengths=True, batch_first=True, fix_length=max_length)
LABEL = torchtext.data.Field(sequential=False, use_vocab=False)


# フォルダ「data」から各tsvファイルを読み込む. テストデータと訓練+バリデーションに分割
train_val_ds, test_ds = torchtext.data.TabularDataset.splits(
    path='./bert-master/JAS/', train='train.tsv',
    test='test.tsv', format='tsv', skip_header=True,
    fields=[('Text', TEXT), ('Label', LABEL)])
```

```
# torchtextで単語ベクトルとして読み込む
from torchtext.vocab import Vectors

japanese_word2vec_vectors = Vectors(
    name='./data/japanese_word2vec_vectors.vec')

# 単語ベクトルの中身を確認
print("1単語を表現する次元数：", japanese_word2vec_vectors.dim)
print("単語数：", len(japanese_word2vec_vectors.itos))
```

```
import numpy as np
import xgboost as xgb
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, log_loss
#　平均
def to_numpy_dataset_average(ds, vocab_vec, stoi):
  return  np.array([ np.average([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) for ex in ds ]), np.array([ int(ex.Label) for ex in ds ]).reshape(len(ds), 1)
# 最大値
def to_numpy_dataset_max(ds, vocab_vec, stoi):
  return  np.array([ np.max([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) for ex in ds ]), np.array([ int(ex.Label) for ex in ds ]).reshape(len(ds), 1)
# 最小値
def to_numpy_dataset_min(ds, vocab_vec, stoi):
  return  np.array([ np.min([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) for ex in ds ]), np.array([ int(ex.Label) for ex in ds ]).reshape(len(ds), 1)

# [平均 ,最大 , 最小値]
def to_numpy_dataset_all(ds, vocab_vec, stoi):
  ave = np.array([ np.average([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) if len(ex.Text) != 0 else np.zeros(200) for ex in ds ])
  mx = np.array([ np.max([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) if len(ex.Text) != 0 else np.zeros(200) for ex in ds ])
  mn = np.array([ np.min([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) if len(ex.Text) != 0 else np.zeros(200) for ex in ds ])
  return np.c_[ ave, mx, mn ], np.array([ int(ex.Label) for ex in ds ]).reshape(len(ds), 1)

def preprocess(ds, TEXT):
  vocab_vec = TEXT.vocab.vectors.numpy()
  return to_numpy_dataset_all(ds, vocab_vec, TEXT.vocab.stoi)

def train(dtrain, dvalid, params):
  watchlist = [(dtrain, 'train'),(dvalid, 'eval')]
  evals_result = {}
  model = xgb.train(params, dtrain, num_boost_round=1000, verbose_eval=1, early_stopping_rounds=10, evals=watchlist, evals_result=evals_result)
  return model, evals_result
```

```
def main():
  scores_accuracy = []
  scores_logloss = []
  params = {
      'booster' : 'gbtree',
      'objective' : 'multi:softprob',
      'eval_metric' : 'mlogloss',
      'eta' : 0.1,
      'gamma' : 0.8,
      'alpha' : 0.0,
      'lambda' : 1.0,
      'min_child_weight' : 1,
      'max_depth' : 5,
      'subsample' : 0.8,
      'colsample_bytree' : 0.8,
      'random_state' : 57,
      'num_class' : 6
  }
  kf = KFold(n_splits=5, shuffle=True, random_state=57)
  TEXT.build_vocab(train_val_ds, vectors=japanese_word2vec_vectors, min_freq=1)
  train_val_x, train_val_y = preprocess(train_val_ds, TEXT)
  for train_idx, val_idx in kf.split(train_val_x, train_val_y):
    tr_x, tr_y = train_val_x[train_idx], train_val_y[train_idx]
    val_x, val_y = train_val_x[val_idx], train_val_y[val_idx]
    dtrain = xgb.DMatrix(tr_x, label=tr_y)
    dvalid = xgb.DMatrix(val_x, label=val_y)

    model, eval_res = train(dtrain, dvalid, params)

    y_pred_proba = model.predict(dvalid)
    y_pred = np.argmax(y_pred_proba, axis=1)
    accuracy = accuracy_score(val_y, y_pred)
    logloss = log_loss(val_y, y_pred_proba)

    scores_accuracy.append(accuracy)
    scores_logloss.append(logloss)

  logloss = np.mean(scores_logloss)
  accuracy = np.mean(scores_accuracy)

  print("logloss: {0}, accuracy: {1}".format(logloss.round(4), accuracy.round(4)))

  return model, eval_res

model, eval_res = main()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# 学習の課程を折れ線グラフとしてプロットする
train_metric = eval_res['train']['mlogloss']
plt.plot(train_metric, label='train mlogloss')
eval_metric = eval_res['eval']['mlogloss']
plt.plot(eval_metric, label='validation mlogloss')
plt.grid()
plt.legend()
plt.xlabel('rounds')
plt.ylabel('mlogloss')
plt.savefig('convergence.png', dpi=300)
```

```
# 評価
test_x, test_y = preprocess(test_ds, TEXT)
dtest = xgb.DMatrix(test_x, label=test_y)
y_pred_proba = model.predict(dtest)
y_pred = np.argmax(y_pred_proba, axis=1)

accuracy = accuracy_score(test_y, y_pred)
logloss = log_loss(test_y, y_pred_proba)

print("logloss: {0}, accuracy: {1}".format(logloss.round(4), accuracy.round(4)))
```

```
# F-score
from sklearn.metrics import f1_score
# one-hot表現へ
onehot_test_y = np.identity(6)[test_y.reshape(len(test_y))]
one_hot_test_pred= np.identity(6)[y_pred]

# mean-f1
mean_f1 = f1_score(onehot_test_y, one_hot_test_pred, average='samples')
# macro-f1
macro_f1 = f1_score(onehot_test_y,  one_hot_test_pred, average='macro')
# micro-f1
micro_f1 = f1_score(onehot_test_y,  one_hot_test_pred, average='micro')

print("mean_f1 : {0}".format(mean_f1))
print("macro_f1 : {0}".format(macro_f1))
print("micro_f1 : {0}".format(micro_f1))
```
