```
import MeCab
import subprocess

cmd='echo `mecab-config --dicdir`"/mecab-ipadic-neologd"'
path = (subprocess.Popen(cmd, stdout=subprocess.PIPE,
                           shell=True).communicate()[0]).decode('utf-8')
m_t = MeCab.Tagger("  -Owakati  -d {0}".format(path))
```

```
def tokenizer_mecab(text):
    text = m_t.parse(text)  # これでスペースで単語が区切られる
    ret = text.strip().split()  # スペース部分で区切ったリストに変換
    return ret

# 前処理として正規化をする関数を定義
import re

def preprocessing_text(text):
    # 改行、半角スペース、全角スペースを削除
    text = re.sub('\r', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('　', '', text)
    text = re.sub(' ', '', text)
    text = re.sub('＿', '', text)

    # 数字文字の一律「0」化
    text = re.sub(r'[0-9 ０-９]', '0', text)  # 数字

    return text


# 前処理とJanomeの単語分割を合わせた関数を定義する

def tokenizer_with_preprocessing(text):
    text = preprocessing_text(text)  # 前処理の正規化
    ret = tokenizer_mecab(text)  # Mecabの単語分割

    return ret
```

```
import torchtext

# tsvやcsvデータを読み込んだときに、読み込んだ内容に対して行う処理を定義
# 文章とラベルの両方に用意します

max_length = 25
TEXT = torchtext.data.Field(sequential=True, tokenize=tokenizer_with_preprocessing,
                            use_vocab=True, lower=True, include_lengths=True, batch_first=True, fix_length=max_length)
LABEL = torchtext.data.Field(sequential=False, use_vocab=False)


# フォルダ「data」から各tsvファイルを読み込む. テストデータと訓練+バリデーションに分割
train_val_ds, test_ds = torchtext.data.TabularDataset.splits(
    path='./bert-master/JAS/', train='train.tsv',
    test='test.tsv', format='tsv', skip_header=True,
    fields=[('Text', TEXT), ('Label', LABEL)])
```

```
import random

# テストデータとバリデーションに分割
train_ds, val_ds = train_val_ds.split(split_ratio=0.8, random_state=random.seed(12))

# 動作確認
print('訓練データの数', len(train_ds))
print('検証データの数', len(val_ds))
print('テストデータの数', len(test_ds))
print('1つ目の訓練データ', vars(train_ds[0]))
```

```
# torchtextで単語ベクトルとして読み込む
from torchtext.vocab import Vectors

japanese_word2vec_vectors = Vectors(
    name='./data/japanese_word2vec_vectors.vec')

# 単語ベクトルの中身を確認
print("1単語を表現する次元数：", japanese_word2vec_vectors.dim)
print("単語数：", len(japanese_word2vec_vectors.itos))
```

```
# 必要な文字だけ辞書として取り出す.
TEXT.build_vocab(train_ds, vectors=japanese_word2vec_vectors, min_freq=1)

print(TEXT.vocab.vectors.shape)

vocab_vec = TEXT.vocab.vectors.numpy()
```

```
# 文章に対応するベクトルを生成(numpy行列として扱う)

#　平均 
def to_numpy_dataset_average(ds, vocab_vec, stoi):
  return  np.array([ np.average([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) for ex in ds ]), np.array([ int(ex.Label) for ex in ds ]).reshape(len(ds), 1)
# 最大値
def to_numpy_dataset_max(ds, vocab_vec, stoi):
  return  np.array([ np.max([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) for ex in ds ]), np.array([ int(ex.Label) for ex in ds ]).reshape(len(ds), 1)
# 最小値
def to_numpy_dataset_min(ds, vocab_vec, stoi):
  return  np.array([ np.min([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) for ex in ds ]), np.array([ int(ex.Label) for ex in ds ]).reshape(len(ds), 1)

# [平均 ,最大 , 最小値]
def to_numpy_dataset_all(ds, vocab_vec, stoi):
  ave = np.array([ np.average([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) for ex in ds ])
  mx = np.array([ np.max([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) for ex in ds ])
  mn = np.array([ np.min([ vocab_vec[stoi[w]] for w in ex.Text ], axis=0) for ex in ds ])
  return np.c_[ ave, mx, mn ], np.array([ int(ex.Label) for ex in ds ]).reshape(len(ds), 1)
```

```
tr_x, tr_y = to_numpy_dataset_all(train_ds, vocab_vec, TEXT.vocab.stoi)
val_x, val_y = to_numpy_dataset_all(val_ds, vocab_vec, TEXT.vocab.stoi)
test_x, test_y = to_numpy_dataset_all(test_ds, vocab_vec, TEXT.vocab.stoi)

print(tr_x.shape, tr_y.shape)
print(val_x.shape, val_y.shape)
print(test_x.shape, test_y.shape)
```

```
# log loss
import xgboost as xgb
from sklearn.metrics import log_loss

dtrain = xgb.DMatrix(tr_x, label=tr_y)
dvalid = xgb.DMatrix(val_x, label=val_y)
dtest = xgb.DMatrix(test_x)

params = { 'objective' : 'multi:softprob' , 'silent' : 1, 'random_state' : 71, 'num_class': 6}
num_round = 50

watchlist = { (dtrain, 'train'), (dvalid, 'eval')}
model = xgb.train(params, dtrain, num_round, evals=watchlist)

va_pred = model.predict(dvalid)
score = log_loss(val_y, va_pred)
print("logloss : {0}".format(round(score,5)))
```

```
# F-score
from sklearn.metrics import f1_score

# one-hot表現へ
onehot_val_y = np.identity(6)[val_y.reshape(len(val_y))]
one_hot_va_pred = np.identity(6)[np.argmax(va_pred, axis=1)]

# mean-f1
mean_f1 = f1_score(onehot_val_y, one_hot_va_pred, average='samples')
# macro-f1
macro_f1 = f1_score(onehot_val_y,  one_hot_va_pred, average='macro')
# micro-f1
micro_f1 = f1_score(onehot_val_y,  one_hot_va_pred, average='micro')

print("mean_f1 : {0}".format(mean_f1))
print("macro_f1 : {0}".format(macro_f1))
print("micro_f1 : {0}".format(micro_f1))
```
